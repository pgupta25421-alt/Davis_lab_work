```python
def classify(num):
    if num > 0:
        print("Positive")
    elif num < 0:
        print("Negative")
    else:
        print("Zero")

    # Even or Odd
    if num != 0:
        if num % 2 == 0:
            print("Even")
        else:
            print("Odd")

numbers = [1, -2, 0, 5]

for n in numbers:
    print("\nNumber:", n)
    classify(n)
```

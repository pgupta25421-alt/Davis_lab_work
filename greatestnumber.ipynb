```python
# Input number
num = int(input("Enter a number: "))

# Convert number to positive (in case of negative input)
num = abs(num)

# Variable to store greatest digit
max_digit = 0

# Loop through digits
while num > 0:
    digit = num % 10        # get last digit
    if digit > max_digit:
        max_digit = digit
    num = num // 10         # remove last digit

print("Greatest digit:", max_digit)
```
